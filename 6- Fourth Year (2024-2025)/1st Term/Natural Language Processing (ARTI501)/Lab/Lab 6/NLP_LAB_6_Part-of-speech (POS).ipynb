# Part-of-speech (POS)

Part-of-speech (POS) tagging is a popular Natural Language Processing technique which refers to categorizing words in a text (corpus) in correspondence with a particular part of speech, depending on the definition of the word and its context.

## Benefits of POS tagging



POS Taggning is mostly usefule in Named Entity Recognition (NER) applications and extracting relations between words.

**POS Tagging can be used in:**
- Grammar correction
- Spelling correction
- Question-Answering
- What else?

## POS tagging using NLTK

### List of POS tags and their meaning in NLTK


**CC:** Coordinating conjunction

**CD:** Cardinal number

**DT:** Determiner

**EX:** Existential there

**FW:** Foreign word

**IN:** Preposition or subordinating conjunction

**JJ:** Adjective

**JJR:** Adjective, comparative

**JJS:** Adjective, superlative

**LS:** List item marker

**MD:** Modal

**NN:** Noun, singular or mass

**NNS:** Noun, plural

**NNP:** Proper noun, singular

**NNPS:** Proper noun, plural

**PDT:** Predeterminer

**POS:** Possessive ending

**PRP:** Personal pronoun

**PRP$:** Possessive pronoun

**RB:** Adverb

**RBR:** Adverb, comparative

**RBS:** Adverb, superlative

**RP:** Particle

**SYM:** Symbol

**TO:** to

**UH:** Interjection

**VB:** Verb, base form

**VBD:** Verb, past tense

**VBG:** Verb, gerund or present participle

**VBN:** Verb, past participle

**VBP:** Verb, non-3rd person singular present

**VBZ:** Verb, 3rd person singular present

**WDT:** Wh-determiner

**WP:** Wh-pronoun

**WP$:** Possessive wh-pronoun

**WRB:** Wh-adverb

### pos_tag method

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize

text = "I received my new phone broken from Amazon."
tokens = word_tokenize(text)

for word in tokens:
    print(f'word: ({word}), part of speech: ({pos_tag([word])[0][1]})')
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(u"The quick brown fox jumped over the lazy dog's back.")
```

* To view the coarse POS tag use `token.pos_`
* To view the fine-grained tag use `token.tag_`
* To view the description of either type of tag use `spacy.explain(tag)`

```python
# Print the full text:
print(doc.text)
```

```python
# Print the fifth word and associated tags:
print(doc[4].text, doc[4].pos_, doc[4].tag_, spacy.explain(doc[4].tag_))
```

```python
for token in doc:
    print(f'{token.text:{10}} {token.pos_:{8}} {token.tag_:{6}} {spacy.explain(token.tag_)}')
```

In the English language, the same string of characters can have different meanings, even within the same sentence. For this reason, morphology is important. **spaCy** uses machine learning algorithms to best predict the use of a token in a sentence. Is *"I read books on NLP"* present or past tense? Is *wind* a verb or a noun?

```python
doc = nlp(u'I read books on NLP.')
r = doc[1]

print(f'{r.text:{10}} {r.pos_:{8}} {r.tag_:{6}} {spacy.explain(r.tag_)}')
```

```python
doc = nlp(u'I am reading a book.')
r = doc[1]

print(f'{r.text:{10}} {r.pos_:{8}} {r.tag_:{6}} {spacy.explain(r.tag_)}')
```

In the first example, with no other cues to work from, spaCy assumed that ***read*** was present tense.<br>In the second example the present tense form would be ***I am reading a book***, so spaCy assigned the past tense.

### Visualizing Parts of Speech
spaCy offers an outstanding visualizer called **displaCy**:

```python
# Perform standard imports
import spacy
nlp = spacy.load('en_core_web_sm')

# Import the displaCy library
from spacy import displacy
```

```python
# Create a simple Doc object
doc = nlp(u"The quick brown fox jumped over the lazy dog's back.")
```

```python
# Render the dependency parse immediately inside Jupyter:
displacy.render(doc, style='dep', jupyter=True, options={'distance': 110})
```

## Example: Find the negetive adjective words in a sentance

This example you can extended later for creating a machine learning model to find the negetive words, extract useful information, find specefic words in a text. 

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
text =  "Mobile is very handy. Packed with good performances and nice camera. This is my first iPhone. Very good experience with Apple. Battery is worst. I am hardly getting 5 hours screen on time. Positive thing is I came to know how many hours I was spending on my mobile. Because of poor battery now I am only using mobile for 4 hours a day."

print(text, end='\n\n')

for sent in sent_tokenize(text):
    tokens = word_tokenize(sent)
    part_of_speech = pos_tag(tokens)
    
    for i in part_of_speech:
        if 'JJ' in i:
            print(i)
    
    print('============================')
```
